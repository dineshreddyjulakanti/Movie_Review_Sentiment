```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentance=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'your videos are good',
]
```

```python
voc_size=10000
```

```python
# one hot encoding
one_hot_rep=[one_hot(words,voc_size)for words in sentance]
```

```python
from  tensorflow.keras.layers import Embedding
# from  tensorflow.keras.processing import pad_sequences
from  tensorflow.keras.utils import pad_sequences
from  tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_doc=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
```

```python
# feature representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
